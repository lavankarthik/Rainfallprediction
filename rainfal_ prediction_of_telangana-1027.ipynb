```python
df["SUBDIVISION"].nunique()
```

```python
group = df.groupby('SUBDIVISION')['YEAR','JAN','FEB','MAR','APR','MAY','JUN','JUL','AUG','SEP','OCT','NOV','DEC']
df=group.get_group(('TELANGANA'))
df
```

```python
df=df.melt(['YEAR']).reset_index()
df
```

```python
df= df[['YEAR','variable','value']].reset_index().sort_values(by=['YEAR','index'])
df
```

```python
df.YEAR.unique()
```

```python
df.columns=['Index','Year','Month','Avg_Rainfall']
```

```python
df
```

```python
Month_map={'JAN':1,'FEB':2,'MAR' :3,'APR':4,'MAY':5,'JUN':6,'JUL':7,'AUG':8,'SEP':9,
   'OCT':10,'NOV':11,'DEC':12}
df['Month']=df['Month'].map(Month_map)
df
```

```python
df.drop(columns="Index",inplace=True)
```

```python
df
```

```python
df.groupby("Year").sum().plot()
plt.show()
```

```python
x=np.asanyarray(df[['Year','Month']]).astype('int')
y=np.asanyarray(df['Avg_Rainfall']).astype('int')
print(x.shape)
print(y.shape)
```

```python
# splitting the dataset into training and testing in the ratio 80:20
from sklearn.model_selection import train_test_split
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.2, random_state=10)
```

```python
#5.Model Selection and Evaluation
```

```python
#Linear Regression
from sklearn.linear_model import LinearRegression
LR = LinearRegression()
LR.fit(x_train,y_train)
```

```python
LR.score(x,y)
```

```python
#Logistic Regression
from sklearn.linear_model import LogisticRegression
lr= LogisticRegression()
lr.fit(x_train,y_train)
```

```python
lr.score(x,y)
```

```python
#Decision Tree
from sklearn.tree import DecisionTreeClassifier
dtc=DecisionTreeClassifier()
dtc.fit(x_train,y_train)
```

```python
dtc.score(x_test,y_test)
```

```python
#Random forest
from sklearn.ensemble import RandomForestRegressor
rfr = RandomForestRegressor()
rfr.fit(x_train, y_train)
```

```python
rfr.score(x,y)
```

```python
rfr.predict([[2016,11]])
```

```python
#6.Create Pickle file
```

```python
#import the model
import pickle
pickle.dump(rfr,open("rfr.pkl","wb"))
```
